# Plotting Analytical Tradeoff Curves for Gaussian and Laplace Mechanisms

## Description 

This notebook demonstrates how to use the API to visualize the analytically computed tradeoff curves for the Gaussian and Laplacian mechanisms in differential privacy.

```python
import numpy as np
from scipy.special import erf
import os
import sys
import matplotlib.pyplot as plt

# Navigate to the parent directory of the project structure
project_dir = os.path.abspath(os.path.join(os.getcwd(), '../..'))
src_dir = os.path.join(project_dir, 'src')

# Add the src directory to sys.path
sys.path.append(src_dir)

from analysis.tradeoff_Gaussian import Gaussian_compute_tradeoff_curve
from analysis.tradeoff_Laplace import Laplace_compute_tradeoff_curve
```

```python
# Generate eta values from 0.01 to 5 (avoid eta=0 to prevent log(0) issues)
eta_values = np.linspace(0.01, 5, 500)
alpha_values, beta_values = Gaussian_compute_tradeoff_curve(eta_values)

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(alpha_values, beta_values, label=r"$\beta(\alpha)$", linewidth=2)
plt.xlabel(r"$\alpha$", fontsize=14)
plt.ylabel(r"$\beta$", fontsize=14)
plt.title(r"Tradeooff Curve for Gaussian pair $N(0,1)$ and $N(1, 1)$", fontsize=16)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```

```python
# Generate eta values from 0.01 to 5 (avoid eta=0 to prevent log(0) issues)
eta_values = np.linspace(0.01, 15, 500)
alpha_values, beta_values = Laplace_compute_tradeoff_curve(eta_values)

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(alpha_values, beta_values, label=r"$\beta(\alpha)$", linewidth=2)
plt.xlabel(r"$\eta$", fontsize=14)
plt.ylabel(r"$\alpha$", fontsize=14)
plt.title(r"Tradeooff Curve for Laplacian pair $L(0,1)$ and $L(1, 1)$", fontsize=16)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```
